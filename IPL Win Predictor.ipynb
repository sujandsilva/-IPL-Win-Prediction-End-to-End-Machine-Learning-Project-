```python
#Label encoding and One hot encoding
encoded_df = pd.get_dummies(df['Toss winner'], prefix='Toss winner')
df_encoded = pd.concat([encoded_df, df], axis=1)
df_encoded = df_encoded.drop(columns=['Toss winner'])

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df_encoded = df_encoded.apply(label_encoder.fit_transform)

pd.DataFrame(df_encoded)
```

```python
#Creating train data and test data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2)
x_test
```

```python
########## The SVC model (rbf kernel) ##########

from sklearn.svm import SVC
classifier_rbf = SVC (kernel = 'rbf', gamma = 2, C=10)
classifier_rbf.fit (x_train,y_train)
```

```python
#The predicted data for the test data
y_pred = classifier_rbf.predict(x_test)
y_pred
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the trained model
model = boostmodel_chosen

# Function to preprocess input data
def preprocess_input(toss_winner, toss_decision, venue):
    # One-hot encoding for toss winner
    toss_winner_encoded = pd.get_dummies(pd.Series(toss_winner)).reindex(columns=teams, fill_value=0)
    
    # Label encoding for toss decision and venue
    toss_decision_encoded = label_encoder_toss_decision.transform([toss_decision])[0]
    venue_encoded = label_encoder_venue.transform([venue])[0]
    
    # Combine encoded features into a single array
    input_features = list(toss_winner_encoded.values.flatten()) + [toss_decision_encoded, venue_encoded]
    
    return [input_features]

# List of teams (assuming the order of teams is consistent)
teams = ['Pune Warriors', 
         'Kolkata Knight Riders',
         'Rajasthan Royals',
         'Kochi Tuskers Kerala',
         'Gujarat Lions',
         'Chennai Super Kings', 
         'Delhi Daredevils', 
         'Deccan Chargers', 
         'Delhi Capitals',
         'Mumbai Indians', 
         'Sunrisers Hyderabad', 
         'Rising Pune Supergiants', 
         'Royal Challengers Bangalore', 
         'Kings XI Punjab']

# Initialize label encoders for toss decision and venue
label_encoder_toss_decision = LabelEncoder()
label_encoder_venue = LabelEncoder()

# Label encoding for toss decision and venue
label_encoder_toss_decision.fit(['bat', 'field'])
label_encoder_venue.fit(['home', 'away'])

# Input from the user
toss_winner = input("""Enter the team that won the toss: 
                       Choices:
                       1. Pune Warriors
                       2. Kolkata Knight Riders
                       3. Rajasthan Royals
                       4. Kochi Tuskers Kerala
                       5. Gujarat Lions
                       6. Chennai Super Kings 
                       7. Delhi Daredevils
                       8. Deccan Chargers
                       9. Delhi Capitals
                       10.Mumbai Indians
                       11.Sunrisers Hyderabad
                       12.Rising Pune Supergiants
                       13.Royal Challengers Bangalore
                       14.Kings XI Punjab 
                       """)
toss_decision = input("Enter the toss decision (bat/field): ")
venue = input("Enter the venue (home/away): ")

# Preprocess input data
input_features = preprocess_input(toss_winner, toss_decision, venue)

# Make prediction
prediction = model.predict(input_features)

# Display the result
if prediction == 1:
    print("The team is predicted to win")
else:
    print("The team is predicted to lose")
```
